```python
#importar las librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
datos = pd.read_csv('vacunas.csv')
```

```python
dinicial = 0
minicial = 0
dfinal = 0
mfinal = 0
```

```python
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

rango1 = []
contador=0
vr1=[]

#print("Ingrese el dia inicial")
#dinicial = input()
#print("Ingrese el mes inicial")
#minicial = input()

#print("Ingrese el dia final")
#dfinal = input()
#print("Ingrese el mes final")
#mfinal = input()

start_date = date(2021, minicial, dinicial)
end_date = date(2021, mfinal, dfinal)



listaD = [list(row) for row in datos.values]

for l in listaD:
    for single_date in daterange(start_date, end_date):
        if l[0] == (single_date.strftime("%d/%m/%Y")):
            #print (single_date.strftime("%Y-%m-%d"))
            rango1.append(listaD[contador])
    contador=contador+1
    
vr1 = np.array(rango1)

      

plt.figure(figsize=(20,12))
plt.title('Vacunas')

plt.plot(vr1[:,0], vr1[:,2], 'b.-')
plt.plot(vr1[:,0], vr1[:,3], 'r.-')
plt.plot(vr1[:,0], vr1[:,1], 'y.-')

plt.xticks(vr1[:,0].tolist(), rotation=60)

plt.xlabel('Fecha')
plt.ylabel('Numero de Dosis')
plt.legend(['Primera Dosis','Segunda Dosis', 'Dosis totales'], loc='lower right')
plt.show()
```
